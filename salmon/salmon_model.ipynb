\begin{array}{l}
\frac{dN}{dt} = -(\mu + h_r)N \\
\frac{dW}{dt} = g_r\left(1 - \frac{W}{W_{max}} \right) - h_r(W-W_{post}) \\ 
\frac{dV}{dt} = \sigma^2_gg_r - h_r(V-V_{post}) \\ 
\end{array}

Harvesting removes salmon with $w \geq 5$, calculated using the cumulative distribution function (CDF) of the normal distribution: $$\Phi(z) = \frac{1}{\sqrt{2\pi}} \int^{z}_{-\infty} e^{\frac{-u^2}{2}}du$$

$$P(w \geq 5) = 1 - \Phi \left( \frac{5-W}{\sqrt{V}} \right) \qquad \overset{\text{harvested rate}}{\longrightarrow} \qquad h_r = h \left[ 1 - \Phi \left( \frac{5-W}{\sqrt{V}} \right) \right]$$

Truncated Gaussian Distribution
For a Gaussian random variable $w \sim \mathcal{N}(\mu, \sigma^2)$, truncated to $w < a$, the probability density function of the truncated distribution is:



$$f(w|w < a) = \frac{\frac{1}{\sqrt{2 \pi \sigma^2}} \exp \left( - \frac{(w-\mu)^2}{2\sigma^2} \right) }{\Phi \left( \frac{a-\mu}{\sigma} \right)} \quad \text{for } w < a$$



where $\Phi(z)$ is the standard normal CDF, and the denominator $\Phi\left(\frac{a - \mu}{\sigma}\right) = P(w < a)$ is the probability of $w < a$.
In our case:

+ $\mu = W(t)$: Mean weight before harvesting.

+ $\sigma^2 = V(t)$: Variance before harvesting.

+ $a=5$: Truncation point (harvest salmon with $w \geq 5$).



Letâ€™s define:

+ $\alpha = \frac{a - \mu}{\sigma} = \frac{5 - W(t)}{\sqrt{V(t)}}$: Standardized truncation point.

+ $\phi(z) = \frac{1}{\sqrt{2\pi}} \exp\left(-\frac{z^2}{2}\right)$: Standard normal PDF.

+ $\Phi(\alpha) = P(z < \alpha)$: CDF at $\alpha$.

The fraction of salmon remaining after harvesting is $\Phi(\alpha)$, and the fraction harvested is $1 - \Phi(\alpha)$.



##### Conditional Mean ($W_{\text{post}}$)
The mean of a Gaussian distribution truncated at $w < a$ is:

$$\mathbb{E}[w|w<a] = \mu - \sigma \frac{\phi \left( \frac{a-\mu}{\sigma}\right)}{\Phi \left( \frac{a-\mu}{\sigma}\right)}$$

Substituting $\mu = W(t)$, $\sigma = \sqrt{V(t)}$,$a = 5$:

$$W_{\text{post}} = W(t) - \sqrt{V(t)} \frac{\phi \left( \frac{5-W(t)}{\sqrt{V(t)}}\right)}{\Phi \left( \frac{5-W(t)}{\sqrt{V(t)}}\right)}$$



Simplify with $\alpha = \frac{5 - W(t)}{\sqrt{V(t)}}$:

$$W_{\text{post}} = W(t) - \sqrt{V(t)} \frac{\phi(\alpha)}{\Phi(\alpha)}$$

where:

+ $\phi(\alpha) = \frac{1}{\sqrt{2\pi}} \exp\left(-\frac{\alpha^2}{2}\right)$

+ $\Phi(\alpha) = \int_{-\infty}^{\alpha} \phi(z) \, dz$

The term $\frac{\phi(\alpha)}{\Phi(\alpha)}$
is the hazard function (or inverse Mills ratio) for the standard normal, which adjusts the mean downward since larger weights are removed.




##### Conditional Variance ($V_{\text{post}}$)

The variance of the truncated Gaussian $w < a$ is:

$$Var[w|w<a] = \sigma^2 \left[ 1 - \frac{\phi \left( \frac{a-\mu}{\sigma}\right)  \frac{a-\mu}{\sigma}}{\Phi \left( \frac{a-\mu}{\sigma}\right)} - \left( \frac{\phi \left( \frac{a-\mu}{\sigma}\right)}{\Phi \left( \frac{a-\mu}{\sigma}\right)} \right)^2 \right]$$

Substituting $\sigma^2 = V(t)$, $\mu = W(t)$, $a=5$, and $\alpha = \frac{5 - W(t)}{\sqrt{V(t)}}$:

$$V_{\text{post}} = V(t) \left[ 1 - \frac{\phi(\alpha) \alpha}{\Phi(\alpha)} - \left( \frac{\phi(\alpha)}{\Phi(\alpha)} \right)^2 \right]$$

This expression accounts for the reduced variance after removing the upper tail of the distribution.



Notes
Simplification: Let $\alpha = \frac{5 - W}{\sqrt{V}}$. Then:

+ $W_{\text{post}} = W - \sqrt{V} \cdot \frac{\phi(\alpha)}{\Phi(\alpha)}$

+ $V_{\text{post}} = V \left[ 1 - \frac{\phi(\alpha) \cdot \alpha}{\Phi(\alpha)} - \left( \frac{\phi(\alpha)}{\Phi(\alpha)} \right)^2 \right]$

Behavior:

+ If $W \ll 5$, $\alpha$ is large, $\Phi(\alpha) \approx 1$, $\phi(\alpha) \approx 0$, $W_{\text{post}} \approx W$,$V_{\text{post}} \approx V$, as few salmon are harvested.

+ If $W \gg 5$, $\alpha$ is negative and large, $\Phi(\alpha) \approx 0$, harvesting dominates, but $W_{\text{post}}$ and $V_{\text{post}}$ stabilize to values reflecting the truncated distribution.

Numerical Stability: Ensure $V > 0$ to avoid division by zero in $\alpha$. In practice, $V_{\text{post}}$ is always positive but smaller than (V).



We want to model even the mean and standard deviation of the harvested fish: 

$$f(w|w \geq a) = \frac{\frac{1}{\sqrt{2 \pi \sigma^2}} \exp \left( - \frac{(w-\mu)^2}{2\sigma^2} \right) }{1 - \Phi \left( \frac{a-\mu}{\sigma} \right)} \quad \text{for } w \geq a$$

We compute the mean of the truncated distribution:

$$\mathbb{E}[w|w \geq a] = \mu + \sigma \frac{\phi \left( \frac{a-\mu}{\sigma}\right)}{1 - \Phi \left( \frac{a-\mu}{\sigma}\right)}$$

In our case:

+ $\mu = W(t)$: Mean weight before harvesting.
+ $\sigma = \sqrt{V(t)}$: Standard deviation of weight.
+ $a = 5$: Harvesting threshold.

Define: $\alpha = \frac{a - \mu}{\sigma} = \frac{5 - W(t)}{\sqrt{V(t)}}$
The mean weight of harvested salmon (W_{\text{harvested}}) is:

$$W_{\text{harvested}} = W(t) + \sqrt{V(t)} \cdot \frac{\phi\left(\frac{5 - W(t)}{\sqrt{V(t)}}\right)}{1 - \Phi\left(\frac{5 - W(t)}{\sqrt{V(t)}}\right)}$$

Using $\alpha$:

$$W_{\text{harvested}} = W(t) + \sqrt{V(t)} \cdot \frac{\phi(\alpha)}{1 - \Phi(\alpha)}$$
